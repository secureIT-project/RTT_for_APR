### PSB example

````python
code = """
#include <vector>
#include <iostream>
#include <string>
#include <cstring>
#include <queue>
#include <stdio.h>
#include <math.h>
#include <map>
#include <set>
#include <stack>
#include <climits>
using namespace std;
/*
Given a number of cents, Ô¨Ånd the fewest number of US coins (pennies, nickles, dimes, quarters) needed to make that amount, and return the number of each type of coin as a separate output.
For example,
input:
1
output:
1
0
0
0
input:
2
output:
2
0
0
0
input:
3
output:
3
0
0
0
input:
4
output:
4
0
0
0
input:
5
output:
0
1
0
0
*/
int main() {
"""

response = openai.ChatCompletion.create(
    model=MODEL,
    messages=[
        {"role": "system", "content": "You are an experienced software developer in C++. " + \
             "The initial code is given in triple quotes. " + \
             "The task description is given between /* and */. Example input-output pairs are given after prefixes 'input:' and 'output:'" + \
             "Complete the main function according to task. Return only code with correct tabulation."},
        {"role": "user", "content": f'"""{code}' + '\n"""'}
    ],
    temperature=0,
)

response
````

# Test round 2

````python
problems = ['bowling', 'coin-sums']

task_description = {}
task_description["bowling"] = "Given a string representing the individual bowls in a 10-frame round of 10 pin bowling, return the score of that round."
task_description["coin-sums"] = "Given a number of cents, Ô¨Ånd the fewest number of US coins (pennies, nickles, dimes, quarters) needed to make that amount, and return the number of each type of coin as a separate output."

temp = {}

temp['C++'] = """
#include <vector>
#include <iostream>
#include <string>
#include <cstring>
#include <queue>
#include <stdio.h>
#include <math.h>
#include <map>
#include <set>
#include <stack>
#include <climits>
using namespace std;
/*
$prompt
*/
int main() {
"""

temp['Python'] = '''
import os
import sys
import numpy as np
import math
import datetime
import collections
import itertools
import queue
import re
"""
$prompt
"""
'''

languages = ['Python', 'C++']
````

````python
from string import Template
code = {}
for problem in problems:
    code[problem] = {}
    for language in languages:
        template = Template(temp[language])
        code[problem][language] = template.substitute(prompt=task_description[problem])
````

## Initial prompt - generate

````python
system = 'You are an experienced $language developer. You will be given the first lines of the program, ' + \
        'a task description and an open main function. ' + \
        'The program must read arbitrary input and return output as described in the task description. ' + \
        'Examples of input are given after "input:", corresponding output is given after "output:", ' + \
        'task description is given in a docstring format after the list libraries. Generate only code.' 

gen_code = {}

for p in problems:
    gen_code[p] = {}
    for l in languages:
        gen_code[p][l] = {}
        for model in ['gpt-3.5-turbo', 'gpt-4']:
            gen_code[p][l][model] = ''
            user = code[p][l]
            response = openai.ChatCompletion.create(
                model=model,
                messages=[
                    {"role": "system", "content": Template(system).substitute(language=l)},
                    {"role": "user", "content": user + '\n Complete the code'}
                    ],
                    temperature=0,
                )
            gen_code[p][l][model] = response["choices"][0]["message"]["content"]
            print('\n\n\n', p, l, model, '\n\n', gen_code[p][l][model])
````

````python
system = 'You are an experienced $language developer. You will be given the first lines of the program, ' + \
        'a task description and an open main function. ' + \
        'The program must read arbitrary input and return output as described in the task description. ' + \
        'Examples of input are given after "input:", corresponding output is given after "output:", ' + \
        'task description is given in a docstring format after the list libraries. Generate only code.' 

gen_code = {}

for p in problems:
    gen_code[p] = {}
    for l in languages:
        gen_code[p][l] = {}
        for model in ['gpt-3.5-turbo', 'gpt-4']:
            gen_code[p][l][model] = ''
            user = task_descr_io_only[p]
            response = openai.ChatCompletion.create(
                model=model,
                messages=[
                    {"role": "system", "content": Template(system).substitute(language=l)},
                    {"role": "user", "content": user + '\n Generate the code'}
                    ],
                    temperature=0,
                )
            gen_code[p][l][model] = response["choices"][0]["message"]["content"]
            print('\n\n\n', p, l, model, '\n\n', gen_code[p][l][model])
````
